**Now we will do some other observations about the numeric columns, ```price``` , ```kilometer``` and ```powerps```, these three columns may containt valueable information of our dataset**


````python
#Create a explore function to clearly display each trails of the series we input
def explore(series):
    unique_values = series.unique().shape[0]
    print('The series "{s}" has {u} unique values'
         .format(s=series.name, u=unique_values))
    print('\n')
    print('Description')
    print(series.describe())
    print('\n')
    print('Value counts top')
    print(series.value_counts()
          .sort_index(ascending=False)
          .head(10))
````

````python
explore(autos['price'])
print('\n')
explore(autos['kilometer'])
print('\n')
explore(autos['powerps'])
````

**It appears that ```price``` and ```powerps``` columns have some irrealistic data. For example the highest price of a car is more than 2 billions and the lowest price for of a car is 0, in addition, in the ```powerps``` column, the highest horsepower of a car reaches 20,000, which is higher than a fast flying fighter jet, while the lowerst horsepower is 0.**

We use bool to change the irrealistic value into np.nan, with a normal price range between 10 - 900,000EU

````python
#Creating an resonable range
autos.loc[(autos['price']<10)|(autos['price']>900000),'price'] = np.nan
````

````python
explore(autos['price'])
````

After cleaning, the data appears to be more representable, with a mean value of 5965 dollars and a highest car price of 849,000EU

Now we will start to clean ```powerps```. We give a range of 70-1500 horsepower, as only few cars have less than 70 horsepower and over 1500 horspower.

````python
autos.loc[(autos['powerps']<70)|(autos['powerps']>1500),'powerps'] = np.nan
````

**After all the data only represent date information, now we will import ```datetmie``` to convert the string into datetype**

````python
import datetime as dt
#Create a function that transfer a whole series into datetime 
date_format = '%Y-%m-%d'
def date_convert(series):
    date_dt = pd.to_datetime(series, format=date_format)
    return date_dt
````

````python
autos["datecrawled"] = date_convert(autos['datecrawled'])
autos["ad_created"] = date_convert(autos['ad_created'])
autos["lastseen"] = date_convert(autos['lastseen'])
````

## Data Aggregation by Vehicle Brand
**In this section, we are going to take a look of weather the price of the vehicle is associated with mileage and brand, and find potential correlaton between them**

We will use ```.value_counts()``` to see the percentage of different brand in this dataset with ```normalize = 'Ture'```

````python
print('Unique brands: ', len(autos['brand'].unique()))
autos['brand'].value_counts(normalize=True)
````

We can see that ```Vokkswagen```, ```bmw```, ```opel```, ```mercedes_benz``` and ```audi``` are the most popular brands on German ebay.

**Now we are tryting to create a dictionary, which content the mean price as the value and brand name as the key**

````python
#Create a dictionary to store brand name and price
brand_mean_price = {}
#use .unique() to get unique brand name in the data set
brands = autos['brand'].unique()
for i in brands:
    brand = autos[autos['brand']==i]
    mean_price = brand['price'].mean()
    brand_mean_price[i] = round(mean_price,2)

dict(sorted(brand_mean_price.items(), key=lambda item: item[1], reverse = True))
````

**Another loop with brand name and kilometer as key value pair**

````python
mean_kilometer = {}
brands = autos['brand'].unique()
for i in brands:
    brand = autos[autos['brand'] == i]
    kilometer = brand["kilometer"].mean()
    mean_kilometer[i] = round(kilometer)

dict(sorted(mean_kilometer.items(), key=lambda item: item[1], reverse = True))
````

**Key value pair of brand name and registration year**

````python
mean_year = {}
brands = autos['brand'].unique()
for i in brands:
    brand = autos[autos['brand'] == i]
    years = brand["registration_year"].mean()
    mean_year[i] = round(years)

dict(sorted(mean_year.items(), key=lambda item: item[1], reverse = True))
````

**Now we will use ```pd.DateFrame``` and ```pd.Seires``` to first transfer the dictionary into series and then into dateframe, with the column name being listed as ```mean_price```, ```mean_kilo``` and ```mean_registration_year```**

````python
mean_price_df = pd.DataFrame(pd.Series(brand_mean_price), columns = ["mean_price"])
mean_kilometer_df = pd.DataFrame(pd.Series(mean_kilometer), columns = ["mean_kilo"])
mean_year_df = pd.DataFrame(pd.Series(mean_year), columns = ["mean_registration_year"])
````

**Merge these three dataframe into one, with ```pd.merge```**

**Use a list to choose those famous brand**

````python
frame_1 = pd.merge(mean_price_df, mean_kilometer_df, left_index=True, right_index=True)
pd.merge(frame_1, mean_year_df, left_index=True, right_index=True).loc[['volkswagen','bmw', 'opel', 'mercedes_benz', 'audi','ford']].sort_values(by = 'mean_price')
````
